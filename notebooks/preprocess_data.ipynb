```python
df_processed = data.groupby('query').agg(lambda x: list(x)).reset_index()
```

```python
df_processed.head(5)
```

```python
def make_data_for_preds(df, save_flg=False):
    df_tmp = df.copy()
    data = []
    for row in df_tmp.iterrows():
        tmp_dict = {}

        query = row[1]["query"]
        
        texts = row[1]["body"]
        
        is_selected = row[1]["label"]
        
        
        tmp_dict["query"] = query
        tmp_dict["passage_text"] = texts
        tmp_dict["is_selected"] = is_selected
        data.append(tmp_dict)
        
    if save_flg:
        json_object = json.dumps(data, ensure_ascii=False)
        
        with open(DATA_DIR_PATH + f'data_for_gen.json', 'w', encoding='utf-8') as f:
            f.write(json_object)

    return data
```

```python
data_for_gen = make_data_for_preds(df_processed, save_flg=True)
```
